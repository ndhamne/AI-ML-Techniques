# Data Preparation and Processing

```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

Creating a simple dataset :
1. An experimental drug was tested on individuals from ages 13 to 100 in a clinical trial
2. The trial had 2100 participants. Half were under 65 years old, half were 65 years or older
3. Around 95% of patients 65 or older experienced side effects
4. Around 95% of patients under 65 experienced no side effects

```python
for i in range(50):
    # The ~5% of younger individuals who did experience side effects
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)
    
    # The ~5% of older individuals who did not experience side effects
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)  

for i in range(1000):
    # The ~95% of younger individuals who did not experience side effects
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    # The ~95% of older individuals who did experience side effects
    random_younger = randint(65,100)
    train_samples.append(random_younger)
    train_labels.append(1)
```

```python
for i in train_samples:
    print(i)
```

```python
for i in train_labels:
    print(i)
```

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
# use shuffle to get rid of any imposed order from data generation process
train_labels, train_samples = shuffle(train_labels, train_samples)
```

```python
# normalise/ standerdize 
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
for i in scaled_train_samples:
    print(i)
```

# Simple tf.keras Sequential Model

```python
# for building the model 
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense

# for training the model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
# if you are runing on GPU uncomment this cell
# physical_devices = tf.config.experimental.list_physical_devices('GPU')
# print("Num GUPs available:", len(physical_devices))
# tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

```python
model = Sequential([
    Dense(units=16, activation='relu', input_shape=(1,)), # Input layer
    Dense(units=32, activation='relu'),                   # Hidden layer
    Dense(units=2, activation='softmax')                  # Output layer
])  # Dense = Fully connected layer
```

```python
model.summary()
```

```python
# prepare the model for training
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# tain the model using fit function:
# x : input data
# y : target data/labels
# batch_size : how many samples included in one batch to be pass in procees at one time
# epochs : model is going to train on all data 30 times before completing training process
# shuffle : default (True)
# verbose : see output (0/1/2) 2: highest level
model.fit(x=scaled_train_samples, y=train_labels, batch_size=10, epochs=30, shuffle=True, verbose=2)
```

# How to create a validation set on the fly during training  

Validation set : Testing the model on unexposed data
Creation: 
- Take a subset of trainig set and separate it into a seaprate set named as validation data.
- Training done on only Training data and model validated on validation data.
- After training the model on the training data, after each epoch the model is tested on the validation data. 
- Model is ONLY learning on training data but the loss and accuracy is calculated on training and validation data.
- Validation data to predict how accurate the model predicts on data that it has not seen before. Giving idea how wel the model genralises
- If the model overfits ?
- Two ways to create validation set for sequential model:
1. To have a completely separate validation set and pass validation data in fit function through validation parameter.
2. No need to explicitly create validation set. kersa does that for us.

```python
model.fit(x=scaled_train_samples, y=train_labels, validation_split=0.1, batch_size=10, epochs=30, shuffle=True, verbose=2)
```

# Using confusion matrix to visualise prediction result from NN during inference

- using confusion matrix we can visulise how well the model did on the test data 
- using scikit-learn to create confusion matrix

```python
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
# function as it is from sklearn
# https://scikit-learn.org/0.18/auto_examples/model_selection/plot_confusion_matrix.html
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
```

```python
cm_plot_labels = ['no_side_effects','had_side_effects']
plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title='Confusion Matrix')
```

The left-to-right diagonal are the correct predictions

Convert images to a format that keras understands :

# Build and Train the CNN

1st layer: Conv2D - accepts image data, 32 filters (arbitrary), k_size - commeon for image data. Layer followed by relu activation layer. Padding is zero i.e. dmensioanlity of images not reduced after convolution operation. Input shape -> implicit input layer. Conv2D is hidden layer. The input layer is the input data itself. (height, width, color_channels)

2nd layer: MaxPool layer -> cut image dimesions in half

3rd layer: Conv2D - input not checked. Filters 64 (arbitrary). General rule of incresing functions as you go into latter layers of the network is comon practice. 

4th layer: Another MaxPool layer

5th layer: Flatten into one dimensional tensor before passing to output layer

6th layer: output layer with softmax activation (to give probabilites for each ocrresponding outputs

```python
model = Sequential([
    Conv2D(filters=32, kernel_size=(3,3), activation='relu', padding='same', input_shape=(224,224,3)),
    MaxPool2D(pool_size=(2,2), strides=2),
    Conv2D(filters=64, kernel_size=(3,3), activation='relu', padding='same'),
    MaxPool2D(pool_size=(2,2), strides=2),
    Flatten(),
    Dense(units=2,activation='softmax'),
])
```

```python
model.summary()
```

Prepare the model for training:

Can use binary_crossentropy (since we have only two possible outputs) instead of categorical_crossnetropy. 

Corresponding Changes to me made: 
    - Output layer: only 1 unit instead of 2
    - sigmoid activation instead of softmax
    
These two approaches are totally equivalent
    

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

Train the model

Note: Do not need to specify 'y' (target data) when data is stored as Generator, it contains the corresponding labels as a part of the data point 

```python
model.fit(x=train_batches , validation_data=valid_batches, epochs=10, verbose=2)
```

```python
# function as it is from sklearn
# https://scikit-learn.org/0.18/auto_examples/model_selection/plot_confusion_matrix.html
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

# Fine tune a pretrained model to classify images
- VGG16 and ImageNet

```python
# Download VGG16 model -
vgg16_model = tf.keras.applications.vgg16.VGG16()
```

```python
vgg16_model.summary()
```

```python
# sanity check
def count_params(model):
    non_trainable_params = np.sum([np.prod(v.get_shape().as_list()) for v in model.non_trainable_weights])
    trainable_params = np.sum([np.prod(v.get_shape().as_list()) for v in model.trainable_weights])
    return {'non_trainable_params': non_trainable_params, 'trainable_params': trainable_params}
```

```python
# sanity check
params = count_params(vgg16_model)
assert params['non_trainable_params'] == 0
assert params['trainable_params'] == 138357544
```

```python
type(vgg16_model)
```

```python
model = Sequential()
for layer in vgg16_model.layers[:-1]: 
    # iterating through all layers except the last one
    model.add(layer)
```

```python
model.summary()
```

```python
# set each layer to not trainable
# layers not to be trained. because the model already has learnt the weights  
for layer in model.layers:
    layer.trainable = False
```

```python
# adding our own output layer 
model.add(Dense(units=2, activation='softmax'))
```

```python
model.summary()
```

```python
params = count_params(model)
assert params['non_trainable_params'] == 134260544
assert params['trainable_params'] == 8194
```
